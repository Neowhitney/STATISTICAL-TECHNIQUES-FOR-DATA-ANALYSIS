```python
#Outlier treatment for bedrooms
plt.boxplot(df.bedrooms)
Q1 = df.bedrooms.quantile(0.25)
Q3 = df.bedrooms.quantile(0.75)
IQR = Q3 - Q1
df = df[(df.bedrooms >= Q1 - 1.5*IQR) & (df.bedrooms <= Q3 + 1.5*IQR)]
```

```python
# outlier treatment for price
plt.boxplot(df.price)
Q1 = df.price.quantile(0.25)
Q3 = df.price.quantile(0.75)
IQR = Q3 - Q1
df = df[(df.price >= Q1 - 1.5*IQR) & (df.price <= Q3 + 1.5*IQR)]
```

```python
# outlier treatment for area
plt.boxplot(df.area)
Q1 = df.area.quantile(0.25)
Q3 = df.area.quantile(0.75)
IQR = Q3 - Q1
df = df[(df.area >= Q1 - 1.5*IQR) & (df.area <= Q3 + 1.5*IQR)]
```

```python
# List of numerical columns to plot
numerical_columns = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'furnishingstatus']
#Repeat through each pair of columns
for i in range (len(numerical_columns)):
    for j in range (i + 1, len(numerical_columns)):
        x_col = numerical_columns[i]
        y_col = numerical_columns[j]
        #Create  scatter plot
        sns.scatterplot(x=x_col, y=y_col, data=df)
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        plt.title(f'SCATTER PLOT OF {x_col} vs {y_col}')
        plt.show()
```
